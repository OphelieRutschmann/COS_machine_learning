```python
# Group the communes by canton and save them into separate -csv files (ex. a ZH.csv file). 

# get an array with the names of each canton
cantons_list = df_clean["Canton"].unique()

# for each canton in the list, choose the subset that corresponds to the canton, and save it as "canton".csv file in the folder "Cantons_datasets"
for canton in cantons_list:
    df_canton = df_clean[df_clean["Canton"] == canton]
    filename = str(canton) + ".csv"
    df_canton.to_csv(filename, index=False)
    
```

```python
# Compute the population density at the canton level and rank the cantons from most dense to least dense
# The population density is calculated as: (total population in a canton) / (total surface area)

# Use the same array with the names of each canton as before (I know you wouldn't need to repeat the line of code, but for clarity I added it again)
cantons_list = df_clean["Canton"].unique()

# Create a data frame with population, surface area and density for each canton (columns will be empty for now)
cantons_df = pd.DataFrame(
    index=cantons_list,
    columns=["Total population", "Total area (km²)", "Density (population/km²)"]
)

# For each canton, calculate the total population, the total surface and the density
# Add this information into the newly created data frame

for canton in cantons_list:
    total_residents = df_clean.loc[df_clean["Canton"] == canton, "Residents"].sum()
    total_area = df_clean.loc[df_clean["Canton"] == canton, "Surface area in km²"].sum()
    densities_1 = total_residents / total_area
    cantons_df.loc[canton, ["Total population", "Total area (km²)", "Density (population/km²)"]] = [total_residents, total_area, densities_1]

# Rank the cantons from most dense to least dense
cantons_df = cantons_df.sort_values("Density (population/km²)", ascending=False)

# Alternative version to calculate densities using groupby and apply 
densities_2 = df_clean.groupby("Canton")[["Residents", "Surface area in km²"]].apply(lambda df: df["Residents"].sum()/df["Surface area in km²"].sum())

# Alternative version to calculate densities using groupby and sum (could also work with .agg("sum"):
sums = df_clean.groupby("Canton")[["Residents", "Surface area in km²"]].sum()
densities_3 = sums["Residents"] / sums["Surface area in km²"]

cantons_df
```

```python
# Compute the number of communes in each canton where more than 50 percent of the population are aged between 20 and 64 years old

# for each canton, filter the data_frame to contain only rows with >50% active population and count the number of rows
# add this information to the cantons_df in a new column called "N communes with >50% active pop."
for canton in cantons_list:
    n_communes = len(df_clean[(df_clean["Canton"] == canton) & (df_clean["20-64 years"] > 50)])
    cantons_df.loc[canton, "N communes with >50% active pop."] = n_communes

cantons_df
```

```python
# Compute the difference between the max and min elevations for each canton. 
# Find the top 5 cantons that have the largest range of elevations

# For this task, I'm directly using the groupby function
elevation_max = df_clean.groupby("Canton")["Elevation"].max()
elevation_min = df_clean.groupby("Canton")["Elevation"].min()
elevation_diff = elevation_max - elevation_min
cantons_df["Elevation difference"] = elevation_diff

# Alternative version using "apply" and a lambda function:
elevation_diff_lambda = df_clean.groupby("Canton")["Elevation"].apply(lambda x: x.max()-x.min())

cantons_df
```

## D. Probabilities
Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais

#### Calculations

We know that P(A|B) = P(A & B) / P(B)

Here, we have: 

- P(A): probability of a commune being from Valais 
- P(B): probability of a commune being over 2000 m
- P(A & B): probability of a commune being from valais and over 2000m 

We can calculate:
- P(A) = number of communes in Valais / total number of communes
- P(B) = # of communes above 2000m / total # of communes
- P(A & B) = number of communes above 2000 and in Valais / total number of communes 

To calculate P(A|B), we can simply do:
- number of communes above 2000 and in valais / number of communes above 2000 m

```python
# Calculations
N_Valais_2000 = len(df_clean[(df_clean["Canton"] == "VS") & (df_clean["Elevation"] > 2000)])
N_2000 = len(df_clean[df_clean["Elevation"] > 2000])

prob = (N_Valais_2000 / N_2000) * 100

print("Probability of a commune is from Valais knowing that it is above 2000m is", prob, "%")
```

## E. Matrices

Define a data frame matrix whose rows correspond to communes and the columns to the cantons. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the commune in row i is in the canton in column j and a 0 otherwise

```python
import numpy as np

# Set the indexes of the data frame as the communes
df_clean = df_clean.set_index("Commune", drop=False)

# Get a variable with all the communes and a variable with all the cantons
commune_list = df_clean.index.unique()
canton_list = df_clean["Canton"].unique()

# Initialize a data frame filled with 0. Rows should correspond to the communes, columns to the cantons
results_df = pd.DataFrame(
    np.zeros((len(commune_list), len(canton_list))),
    index = commune_list, 
    columns = canton_list)

# set the value of i,j to 1 if the commune in row i is in the canton in row j
for commune in commune_list:
    canton = df_clean.loc[commune, "Canton"]
    results_df.loc[commune, canton] = 1
    
results_df
```
