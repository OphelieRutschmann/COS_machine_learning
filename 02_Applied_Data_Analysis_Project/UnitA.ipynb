#### **3) Creating columns for each data type**
Here, we will create a variable containing the column names for all of the columns that contain either floats, ints, or objects.

```python
# Return column names for each data type
col_float = df.select_dtypes('float64').columns
col_int = df.select_dtypes('int64').columns
col_obj = df.select_dtypes('O').columns

# Another way to select columns:
col_num = df.loc[:,df.apply(lambda x: pd.api.types.is_numeric_dtype(x))].columns
```
