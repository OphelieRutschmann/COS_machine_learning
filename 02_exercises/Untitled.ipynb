```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import random

%matplotlib inline
```

```python
poisson1 = stats.poisson.rvs(mu=55, size=200000)
poisson2 = stats.poisson.rvs(mu=10, size=100000)
population = np.concatenate((poisson1, poisson2))
```

```python
sample1 = np.random.choice(population, size=1000)
```

```python
t_statistic, p_value = stats.ttest_1samp(sample1, popmean=population.mean())
t_statistic
p_value
```

```python
data = pd.DataFrame(
    ["red"] * 50000 + ["blue"] * 30000 + ["green"] * 10000 + ["white"] * 10000
)
```

```python
sample2 = data.sample(n=1000)
```

```python
data_count = pd.crosstab(index=data[0], columns="count")
sample_count = pd.crosstab(index=sample2[0], columns="count")
```

```python
expected_count = data_count * len(sample2) / len(data)
expected_count
```

```python
chi_square, p_value = stats.chisquare(sample_count,expected_count)
p_value
```
